Q1: Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.



```python
import numpy as np
from scipy.stats import pearsonr

# Sample data
study_time = [10, 12, 8, 15, 7]
exam_scores = [85, 90, 78, 92, 75]

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(study_time, exam_scores)

print(f"Pearson Correlation Coefficient: {pearson_corr}")
```

Interpretation:

If the Pearson correlation coefficient is close to 1, it indicates a strong positive linear relationship (as study time increases, exam scores tend to increase). If it's close to -1, it indicates a strong negative linear relationship. If it's close to 0, there is a weak linear relationship.

Q2: Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.



```python
from scipy.stats import spearmanr

# Sample data
sleep_hours = [7, 6, 8, 5, 7]
job_satisfaction = [8, 6, 9, 4, 7]

# Calculate Spearman's rank correlation
spearman_corr, _ = spearmanr(sleep_hours, job_satisfaction)

print(f"Spearman's Rank Correlation: {spearman_corr}")
```

Interpretation:

Spearman's rank correlation ranges from -1 to 1. A positive value indicates a monotonic increase (as one variable increases, the other tends to increase), while a negative value indicates a monotonic decrease. A value of 0 suggests no monotonic relationship.

Q3: Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.



```python
# Sample data
exercise_hours = [3, 5, 2, 7, 4]
bmi = [25, 30, 22, 35, 28]

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(exercise_hours, bmi)

# Calculate Spearman's rank correlation
spearman_corr, _ = spearmanr(exercise_hours, bmi)

print(f"Pearson Correlation Coefficient: {pearson_corr}")
print(f"Spearman's Rank Correlation: {spearman_corr}")
```

Comparison:

If the Pearson correlation coefficient is high and positive, it indicates a strong linear relationship. However, if the relationship is more nonlinear or there are outliers, Spearman's rank correlation may be more appropriate. A significant difference between the two may suggest the presence of outliers or nonlinearity.

Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below:

The data you provided shows the relationship between age and soft drink preference for a sample of individuals. To analyze the relationship, you can use descriptive statistics and explore patterns in the data. However, it's important to note that with a small dataset like this, you may not draw strong statistical conclusions, and the analysis would be more exploratory in nature.

Here's what you can do:

Descriptive Statistics:

Calculate measures of central tendency (mean, median, mode) for the age variable.
Determine the frequency distribution or mode for soft drink preferences.
Visualization:

Create a bar chart or pie chart to visually represent the distribution of soft drink preferences.
Observations and Conclusions:

Look for any patterns or trends in the data. For example, is there a common preference among certain age groups?
Check if there are any outliers or unusual observations.
Here's a simple Python code snippet using pandas for descriptive statistics and visualization:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Age': [25, 42, 37, 19, 31, 28],
        'Soft Drink Preference': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']}

df = pd.DataFrame(data)

# Descriptive statistics
age_stats = df['Age'].describe()
soft_drink_mode = df['Soft Drink Preference'].mode()

# Visualization
plt.figure(figsize=(8, 6))
df['Soft Drink Preference'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Soft Drink Preference Distribution')
plt.xlabel('Soft Drink')
plt.ylabel('Frequency')
plt.show()

print("Descriptive Statistics for Age:")
print(age_stats)
print("\nMode of Soft Drink Preference:")
print(soft_drink_mode)
```

Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample of
30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

```python
import numpy as np
from scipy.stats import pearsonr

# Sample data
sales_calls_per_day = [20, 25, 18, 30, 22, 28, 15, 24, 19, 23, 27, 21, 26, 17, 29, 20, 25, 18, 30, 22, 28, 15, 24, 19, 23, 27, 21, 26, 17, 29]
sales_per_week = [5, 7, 4, 8, 6, 7, 3, 6, 5, 6, 7, 5, 8, 4, 7, 5, 7, 4, 8, 6, 7, 3, 6, 5, 6, 7, 5, 8, 4, 7]  # Corrected length

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(sales_calls_per_day, sales_per_week)

print(f"Pearson Correlation Coefficient: {pearson_corr}")
```

The interpretation of the Pearson correlation coefficient will indicate the strength and direction of the linear relationship between the number of sales calls made per day and the number of sales made per week. If the coefficient is close to 1, it suggests a strong positive linear relationship, while a coefficient close to -1 indicates a strong negative linear relationship. A coefficient close to 0 suggests a weak or no linear relationship.
