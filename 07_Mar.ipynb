Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

Range: The difference between the maximum and minimum values.
Variance: The average of the squared differences from the mean.
Standard Deviation: The square root of the variance.
These measures quantify the amount of spread or dispersion in a dataset. A larger range, variance, or standard deviation indicates greater variability. For example, if you have two datasets with the same mean but different standard deviations, the one with the larger standard deviation has more variability.

Q6. What is a Venn diagram?

Q8. What do you understand about skewness in data?

Skewness measures the asymmetry of a probability distribution. If a distribution is symmetric, it has a skewness of 0. Positive skewness indicates a longer or fatter right tail, while negative skewness indicates a longer or fatter left tail.

Q9. If a data is right skewed then what will be the position of median with respect to mean?

In a right-skewed distribution, the mean is typically greater than the median. The tail on the right side (larger values) pulls the mean in that direction.

Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

Covariance measures how two variables change together, while correlation standardizes this measure, making it easier to interpret. Covariance can take any value, while correlation is normalized to the range [-1, 1]. Both are used in statistical analysis to understand relationships between variables.

Q12. For a normal distribution data what is the relationship between its measure of central tendency?

For a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.

Q13. How is covariance different from correlation?

Covariance measures the degree to which two variables change together, while correlation standardizes this measure by dividing it by the product of the standard deviations of the variables. Correlation is dimensionless and ranges from -1 to 1.

Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

Outliers can significantly impact the mean and standard deviation. The mean is sensitive to extreme values, pulling it towards the outlier. The standard deviation, being influenced by the mean, can also be inflated. The median is more robust in the presence of outliers.

Example: In the dataset [1, 2, 3, 4, 100], the mean is influenced by the outlier 100, resulting in a mean of 22. The median, however, remains 3, providing a more accurate measure of central tendency.
