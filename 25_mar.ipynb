1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?

To load and examine the dimensions of the dataset, you would typically use a programming language like Python with libraries such as Pandas. Here's an example in Python:

```python
import pandas as pd

# Load the dataset
flight_data = pd.read_excel('flight_price.xlsx')  # Replace 'flight_price_dataset.csv' with your actual dataset file name

# Display the dimensions
rows, columns = flight_data.shape
print(f"The dataset has {rows} rows and {columns} columns.")
```

Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.



```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(flight_data['Price'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?



```python
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()

print(f"The range of prices is from {min_price} to {max_price}.")
```

Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.



```python
import seaborn as sns

# Create a boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=flight_data)
plt.title('Flight Prices by Airline')
plt.xticks(rotation=45)
plt.show()
```

Q9. Load the Google Playstore dataset and examine its dimensions. How many rows and columns does the dataset have?

To load and examine the dimensions of the dataset, you can use the following Python code:

```python
import pandas as pd

# Load the dataset
playstore_data = pd.read_csv('data/google_cleaned.csv')  # Replace 'google_playstore_dataset.csv' with your actual dataset file name

# Display the dimensions
rows, columns = playstore_data.shape
print(f"The dataset has {rows} rows and {columns} columns.")
```

Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Rating', data=playstore_data)
plt.title('App Ratings by Category')
plt.xticks(rotation=90)
plt.show()
```

Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.



```python
# Check for missing values
missing_values = playstore_data.isnull().sum()

print("Missing values in the dataset:")
print(missing_values)
```

Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.



```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(playstore_data['Size'], playstore_data['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size')
plt.ylabel('Rating')
plt.show()
```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.



```python
# Create a bar chart
avg_price_by_type = playstore_data.groupby('Type')['Price'].mean()

plt.bar(avg_price_by_type.index, avg_price_by_type.values, color=['blue', 'orange'])
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
playstore_data = pd.read_csv('data/google_cleaned.csv')

# Check the data types of the 'Installs' column
print("Data types of 'Installs' column:")
print(playstore_data['Installs'].dtype)

# Convert 'Installs' to string and then to numeric values
playstore_data['Installs'] = playstore_data['Installs'].astype(str).str.replace(',', '')
playstore_data['Installs'] = pd.to_numeric(playstore_data['Installs'], errors='coerce')  # 'coerce' to handle potential conversion errors

# Check the data types again
print("Updated data types of 'Installs' column:")
print(playstore_data['Installs'].dtype)

# Top 10 most popular apps
top_10_apps = playstore_data.nlargest(10, 'Installs')[['App', 'Installs']]
print("Top 10 most popular apps:")
print(top_10_apps)
```

Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this?

To identify the most popular app categories, you can follow these steps:

Calculate the total installs for each app category.
Visualize the total installs for each category, for example, using a bar chart or pie chart.
Identify the top app categories with the highest total installs.

```python
# Calculate total installs by app category
total_installs_by_category = playstore_data.groupby('Category')['Installs'].sum()

# Sort categories by total installs
sorted_categories = total_installs_by_category.sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(12, 6))
sorted_categories.plot(kind='bar', color='skyblue')
plt.title('Total Installs by App Category')
plt.xlabel('App Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=90)
plt.show()
```

Total Installs by Category: Calculate the total installs for each app category.
Average Rating by Category: Analyze the average user ratings for apps in each category.

Installs: To understand the popularity of each category.
Rating: To gauge user satisfaction with apps in each category.
Number of Apps: To see the competition in each category.
Visualizations:

Bar chart: Displaying total installs by category.
Boxplot or bar chart: Showing the average ratings by category.
