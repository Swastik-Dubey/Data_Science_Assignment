
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

Probability Mass Function (PMF): It is a function that gives the probability of discrete random variables taking on a specific value. For example, in the case of a fair six-sided die, the PMF would assign a probability of 1/6 to each of the possible outcomes (1, 2, 3, 4, 5, 6).

Probability Density Function (PDF): It is a function that describes the likelihood of a continuous random variable falling within a particular range. Unlike the PMF, the PDF represents a range of values rather than specific points. An example is the normal distribution, where the PDF is the bell-shaped curve that describes the distribution of continuous data.

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Cumulative Density Function (CDF): It gives the probability that a random variable will take a value less than or equal to a specific point. It is the cumulative sum of the probabilities up to that point. The CDF is denot by F(x).

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Central Limit Theorem (CLT): It states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution.

Significance: The CLT is crucial in statistics because it allows us to make statistical inferences about a population based on a sample. It enables the use of normal distribution properties even when the original data may not be normally distributed.

Q10: State the assumptions of the Central Limit Theorem.

Independence: The random variables in the sample must be independent of each other.
Sample Size: The sample size should be sufficiently large. While there is no strict rule, a sample size of 30 or more is often considered sufficient.
Identically Distributed: The random variables should be identically distributed, meaning they have the same mean and standard deviation.
